Zadanie 11

```python
column_names  = iris.columns

plt.hist(iris[column_names[0]],width=0.2,color= 'red',label="sepal length")
plt.hist(iris[column_names[1]],width=0.2,color= 'blue',label="sepal width")
plt.hist(iris[column_names[2]],width=0.2,color= 'green',label="petal length")
plt.hist(iris[column_names[3]],width=0.2,color= 'yellow',label="petal width")
plt.legend()
plt.xlabel("Section")
plt.ylabel("Length/width (cm)")
plt.title("Histogram")
plt.show()
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(15,10))
ax1.hist(iris[column_names[0]],width=0.3,color= 'red',label="sepal length")
ax1.set_xlabel("Section")
ax1.set_ylabel("Length (cm)")
ax1.set_title("sepal length")
ax2.hist(iris[column_names[1]],width=0.2,color= 'blue',label="sepal width")
ax2.set_xlabel("Section")
ax2.set_ylabel("Width (cm)")
ax2.set_title("sepal width")
ax3.hist(iris[column_names[2]],width=0.45,color= 'green',label="petal length")
ax3.set_xlabel("Section")
ax3.set_ylabel("Length (cm)")
ax3.set_title("petal length")
ax4.hist(iris[column_names[3]],width=0.2,color= 'yellow',label="petal width")
ax4.set_xlabel("Section")
ax4.set_ylabel("Width (cm)")
ax4.set_title("petal width")
fig.suptitle("Histograms for each class")
```
